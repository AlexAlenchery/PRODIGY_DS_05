```python
# Plotting contributing factors: Light Conditions.
plt.figure(figsize=(10, 6))
sns.countplot(x='Light_Conditions', data=data, palette='muted')
plt.title('Accidents by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plotting contributing factors: Road Surface Conditions.
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Surface_Conditions', data=data, palette='pastel')
plt.title('Accidents by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plotting contributing factors: Weather Conditions.
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Conditions', data=data, palette='dark')
plt.title('Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Data for accident severity distribution.
severity_count = df['Accident_Severity'].value_counts()

# Creating a pie chart.
plt.figure(figsize=(8, 8))
plt.pie(severity_count, labels=severity_count.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Accident Severity Distribution')
plt.show()
```

Inference:
The generalisations and inferences that we can draw from this analysis is that majority accidents do take place in normal conditions such as fine weather with no high winds, daylight and dry road surfaces however the severity is much higher when unnatural conditions are involved such as rain and high winds, wet or snowy road surfaces and darkness or no lighting.
